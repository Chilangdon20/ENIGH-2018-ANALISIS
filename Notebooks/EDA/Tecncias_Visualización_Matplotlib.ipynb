1. Scatter Plot - Gráfico de dispersión.

Los diagramas de dispersión son engañosamente simples y de uso común, ¡pero simples no significa que no sean útiles!

En un diagrama de dispersión, los puntos de datos se representan individualmente con un punto, círculo o alguna otra forma. Estos gráficos son excelentes para mostrar la **relación entre dos variables**, ya que podemos ver directamente la distribución sin procesar  los datos.


Para crear un diagrama de dispersión en Matplotlib, simplemente podemos usar el método `scatter`. Veamos cómo creando un diagrama de dispersión con puntos de datos generados aleatoriamente de muchos colores y tamaños.

Primero, generemos algunos puntos de datos aleatorios, xey, y establezcamos valores aleatorios para colores y tamaños porque queremos una gráfica bonita:


```
import numpy as np
import matplotlib.pyplot as plt

x = np.random.randn(100)
y = np.random.randn(100)
colors = np.random.randn(100)
size = 1000 * np.random.rand(100)
```

Ahora que tenemos nuestras dos variables (x, y) y los colores y tamaños de los puntos, podemos llamar al método `scatter` así:

```
plt.scatter(x,y,c=colors,s=size,alpha=0.2,cmap='viridis')
```

* `alpha = 0.2` El parámetro de entrada alpha opcional (opcional) del método de dispersión nos permite **ajustar el nivel de transparencia para que podamos ver puntos de datos superpuestos.** Comprenderá mejor este concepto, visualmente, en el resultado final.

* `cmap='viridis'` Matplotlib proporciona una amplia gama de mapas de colores, un parámetro opcional, para crear gráficos visualmente atractivos. Por ejemplo, puede establecer `cmap = plt.cm.Blues`, `cmap = plt.cm.autumn` o `cmap = plt.cm.gist_earth`.

El argumento de color, c, se asigna automáticamente a una escala de colores y el argumento de tamaño, s, está en píxeles. Para ver la escala de colores junto al gráfico en el lado derecho, podemos usar el comando colobar ():

```
plt.colorbar()
```

Ahora pongámoslo todo junto y ejecutemos el código completo para ver el resultado. 

```
# 1. Importar modulos
import matplotlib.pyplot as plt
import numpy as np

# 2. Generar aalgunos datos aleatorios
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

# 3. Cree una gráfica de dispersión usando el mapa de colores.
_ = plt.scatter(x, y, c=colors, s=sizes, alpha=0.2,cmap='viridis')
_ = plt.colorbar();  # Para mostrar la escala de colores

```

Podemos ver que este diagrama de dispersión nos ha dado la capacidad de explorar simultáneamente cuatro dimensiones diferentes de los datos. Si se trataba de un conjunto de datos reales, la ubicación x, y de cada punto se puede utilizar para representar dos características distintas.

El tamaño de los puntos se puede usar para representar una tercera característica, y el mapeo de colores se puede usar para representar diferentes clases o grupos de puntos de datos.

2. Lineplots - Gráficos de Linea.

Un gráfico de líneas muestra puntos de datos conectados por segmentos de línea recta, en lugar de mostrarlos individualmente. Este tipo de gráfico es útil para encontrar relaciones entre conjuntos de datos.

Este tipo de gráfico es útil para encontrar relaciones entre conjuntos de datos. Podemos ver fácilmente si los valores de un conjunto de datos se ven afectados por las variaciones en el otro.

**Supongamos que tenemos un conjunto de datos con la población de dos ciudades a lo largo del tiempo y queremos ver si existe alguna correlación en el tamaño de su población.**

 Podemos usar la función `plt.plot()` para graficar la población contra el tiempo. Pondríamos el tiempo en el eje xy la población en el eje y.

 Para comparar los tamaños de población de las dos ciudades, queremos que los datos de población de ambas estén representados en la misma gráfica. Para crear una sola figura con varias líneas, podemos simplemente llamar a la función `plot` varias veces. Y luego, para distinguir entre los dos, podemos ajustar los colores usando la palabra clave color.

 

```
plt.title("Algun titulo cool para el grafico")
plt.xlabel("alguna etiqueta para x-axis")
plt.ylabel("Alguna etiqueta para y-axis");
```

Veamos esto de principio a fin:

```
# Generando los datos
año = [1970,1980,1990,2000,2010,2020]
poblacion_A = [4.9, 5.3, 7.1, 10.7, 13.5, 15.6]
poblacion_B = [44.4, 55.6, 69.7, 87.1, 95.4, 100.5]

# visualizando los datois

_ = plt.plot(año,poblacion_A,color='g')
_ = plt.plot(año,poblacion_B,color='r')
_ = plt.xlabel('Años')
_ = plt.ylabel('Población en millones')
_ = plt.title('Población al paso del tiempo')

plt.show()
```

3. Histogramas.

Los histogramas son útiles para comprender la distribución de puntos de datos. Básicamente, un histograma es un gráfico en el que los datos se dividen en contenedores. Un contenedor es un rango de valores o intervalos. Para cada contenedor, obtenemos el recuento de cuántos valores caen en él. 

El eje **x representa los rangos de intervalos, mientras que el eje y muestra la frecuencia.** Los contenedores generalmente se especifican como intervalos consecutivos que no se superponen de una variable. 

Podemos usar histogramas usando la función `hist()`:

```
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(1000)
_ = plt.hist(x=data);
```

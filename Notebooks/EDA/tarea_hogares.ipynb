```python
columns_names = concentrado.columns
print(columns_names)
```

```python
@interact
def moda(column = columns_names):
    # Seleccionador de columna
    column = column

    print("La moda de la variable" + " " + str(column)+ " "+ "es:")
    print(concentrado[column].value_counts().idxmax())
    print("El valor de la moda es:")
    print(concentrado[column].value_counts().max())
    n =concentrado.shape[0]
    frecuencia_rel_moda = concentrado[column].value_counts().max()
    print("El valor más común corresponde al")
    print(frecuencia_rel_moda/n )
    plt.xticks(rotation=30)
    sns.distplot(concentrado[column], bins=20, kde=True)    
```
